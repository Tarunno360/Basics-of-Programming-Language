```python
#task7
def sorting_bubble(my_list):
    for i in range(0, len(my_list)-1):
        for j in range(0, len(my_list)-i-1):
            if my_list[j]>my_list[j + 1]:
                temp = my_list[j]
                my_list[j]= my_list[j+1]
                my_list[j + 1]= temp

    return my_list

list_1 = input("Please enter your list:").split(",")
list_2 = input("Please enter your list:").split(",")
list_3 = list_1 + list_2
list_4 = []

for i in list_3:
    list_4.append(int(i))

sorting_bubble(list_4)

if len(list_4) % 2 != 0:
    median = list_4[len(list_4)//2]
else:
    median = (list_4[len(list_4)//2 - 1] + list_4[len(list_4)//2])/2


print("Sorted list=", list_4,"\nMedian=", median)
```
