Write a Python program that reads a number and finds the sum of the series of 1 +11 + 111 + 1111 + ….+N terms.

You need to take input using take_input() function, generate the sequence using gen_seq(n) function and print the sum using print_sum(n) function

Sample Input1:

5

Sample Output1:

1 + 11 + 111 + 1111 + 11111

The Sum is: 1234

Sample Input2:

8

```python
def take_input():
   n= int(input())
   return n

def gen_seq(n):
  s1=""
  for i in range (1,n+1):
    s1+= "1"*i
    s1+= "+"
  s1=s1[0:-1]
  print(s1)

def print_sum(n):
  sum=0
  s1=""
  for i in range (1, n+1):
    s1= "1"*i
    sum+= int(s1)
    #print(sum)
  print("The Sum is:", sum)

#driver code
n= take_input()
gen_seq(n)
print_sum(n)
```

```python
#bubble Descending
def bubble_sort(list_1):

  for i in range(0, len(list_1)-1):
    for j in range(0, len(list_1)-1):
      if(list_1[j+1]>list_1[j]):
        list_1[j],list_1[j+1]= list_1[j+1],list_1[j]

  return list_1


#Driver Code
list_1= [-4,24,12,3,-50,14,23,2]
print(bubble_sort(list_1))
```
