```python
#task-11
def rem_duplicate(x):
  l1=[]
  for i in x:
    if i not in l1:
      l1.append(i)
  return tuple(l1)
rem_duplicate(("Hi", 1, 2, 3, 3, "Hi",'a', 'a', [1,2]))
```

```python
#task-13
def function_name(x,y,z):
  if x=='+':
    return y+z
  elif x=='-':
    return y-z
  elif x=='/':
    return y/z
  elif x=='*':
    return y*z
n1=input('Enter operator : ')
n2=float(input('First number : '))
n3=float(input('second number : '))
print(function_name(n1,n2,n3))
```

```python
#task-14
def function_name(sentence,position):
  s1=''
  s2=''
  for i in range(len(sentence)):
    if i%position!=0:
      s1+=sentence[i]
    else:
      s2+=sentence[i]
  s12=sentence[0]+s1
  s3=''
  for i in s2:
    if i!=sentence[0]:
      s3+=i
  return s12+s3
sentence=input('Add sentence : ')
position=int(input('Add position : '))
print(function_name(sentence,position)) 
```

```python
#14
def rem(sentence,position):
    new = sentence[0]
    removed = ''
    for i in range(len(sentence)):         
        if i % position != 0 :
            new += sentence[i]
        else:
            removed += sentence[i]

    return (new + removed[1:])
sentence = input('Sentence: ')
position = int(input('Position: '))

print(rem(sentence,position))   
```
