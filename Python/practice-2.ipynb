Ask user to input their age (integer) and whether he is a government officer or not (boolean). A person is allowed to have pension if he
is a government officer and aged more than 59 years. He will receive pension in future if he is a government officer but not yet more
than 59 years old. He will not receive any pension if he is not a government officer. Now write the program in such a way that tells the
user whether he is eligible for pension right now, or in future or not eligible at all. If he is a freedom fighter he will get 30% more
pension



```python
age = int(input('Please type your age:'))
govt = input('Are you a government Officer? yes or no:')
if (govt=="yes"):
  govt=True
elif(govt=='no'):
  govt=False
else:
  print('invalid input')
freedom = input('Are you a freedomfighter? yes or no:')
if (freedom=='yes'):
  freedom=True
elif(freedom=='no'):
  freedom=False
else:
  print('invalid input')
if (govt==True and age>59 and freedom==True):
  print('30% extra pension now')
elif (govt==True and age<=59 and freedom==True):
  print('30% more pension in future')
elif (govt==True and age>59 and freedom==False):
  print('exact ammount now')
elif (govt==True and age<=59 and freedom==False):
  print('exact ammount in future')
else:
  print('No pension')
```

```python
def show_palindrome(x):
  c=1
  s1=''
  while c<=x:
    s1+=str(c)
    c+=1
  while x>1:
    x-=1
    s1+=str(x)
  return s1
n=int(input('Add bro'))
print(show_palindrome(n))

```
